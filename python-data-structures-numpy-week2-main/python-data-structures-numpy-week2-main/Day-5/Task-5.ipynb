```python
import numpy as np

exam_scores = np.array([72, 85, 90, 66, 88, 95, 79])

print("Average score:", np.mean(exam_scores))
print("Highest score:", np.max(exam_scores))
print("Lowest score:", np.min(exam_scores))
print("Median score:", np.median(exam_scores))

# Purpose: Applies statistical functions to real-world data.
```
