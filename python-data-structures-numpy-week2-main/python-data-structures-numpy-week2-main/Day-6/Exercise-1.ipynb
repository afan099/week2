```python
# Matrix Addition

import numpy as np
 
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = A + B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Addition Result:\n", result)

# Matrix Substraction

result = A - B

print("Subtraction Result:\n", result)

# Matrix Multiplication

result = np.dot(A, B)

print("Matrix Multiplication Result:\n", result)
```
